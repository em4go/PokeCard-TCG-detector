```python
# Get all files in the downloaded_images folder
cards = set(os.listdir("downloaded_images"))
print(cards)
```

```python
with open('cards.json', 'r', encoding="UTF-8") as f:
    cards_json = json.load(f)
```

```python
msgpack_data = msgpack.packb(cards_json)
```

```python
%time
with open('cards.msgpack', 'wb') as f:
    f.write(msgpack_data)
```

```python
%time
with open('cards.msgpack', 'rb') as f:
    msgpack_data = f.read()
    deserialized_data = msgpack.unpackb(msgpack_data, raw=False)
```
